```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset.csv')

x = np.array(data['X'])
y = np.array(data['Y'])
m = len(data)
```

```python
plt.scatter(x,y)
plt.xlabel('Population of City in 10,000s')
plt.ylabel("Profit in $10,000")
plt.show()
```

```python
# kita buat x_0 = 1
x_0 = np.ones(m)
x = np.stack((x_0, x))

# inisialisasi theta bisa berapa aja tpi kita 0 aja
theta = np.array([0,0], dtype=float)
```

```python
def compute_cost(x, y, theta): 
    return 1/(2*m) * np.sum(np.square(theta.dot(x) - y))

def gradient_descent(x, y, theta, iteration, alpha):
    
    m = len(y)
    cost_history = []
    for _ in range(iteration): 
        temp_0 = theta[0] - alpha * 1/m * np.sum((theta.dot(x) - y) * x[0])
        temp_1 = theta[1] - alpha * 1/m * np.sum((theta.dot(x) - y) * x[1]) 
        theta[0] = temp_0
        theta[1] = temp_1
        
        
        cost_history.append(compute_cost(x, y, theta))
        
  

    
    return theta, cost_history

if __name__ == "__main__": 
    fig, ax = plt.subplots(figsize=(12,8))
    
    iteration = 1500
    alpha = 0.01
    
    theta, cost_history = gradient_descent(x, y, theta, iteration, alpha)
    fig = plt.figure(1)
    fig.set_size_inches(20, 15)
    fig.subplots_adjust(top=0.8)
    ax1 = fig.add_subplot(111)
    ax1.set_ylabel('Integral Result')
    ax1.set_xlabel('N case')
    
    ax1.set_title("Monte Carlo result (1)")
    
    line, = ax1.plot([x for x in range(iteration)], cost_history, label="exact integral")


    ax1.legend()


    fig = plt.figure(2)
    fig.set_size_inches(20, 15)
    fig.subplots_adjust(top=0.8)
    ax2 = fig.add_subplot(111)
    ax2.set_ylabel('Integral Result')
    ax2.set_xlabel('N case')
    
    ax2.set_title("Monte Carlo result (1)")
    
    line, = ax1.scatter(data['X'],data['Y'], label="exact integral")


    ax2.legend()



        
```
